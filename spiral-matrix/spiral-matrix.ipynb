The code I wrote below takes as an input a number, n, and builds a matrix of a size of nxn, filled with numbers from 1 to n^2 in a spiral-like manner, i.e., the spiral starts at the top left corner of the matrix and then turns clockwise every time it reaches a new corner.

Take a look at how the code works!

```python
#n=int(input())
n=11
a=n*2-1

c=0
b = [[0 for j in range(n)] for i in range(n)]   #creating a list of lists with zeros occupying the positions of the elemnts in the matrix of size of nxn 
z=1  #assigning the first value to the variable for the numbers in the spiral
if n%2!=0:
    for p in range (0,a//4):
        #filling the rows of the spiral matrix from the top to the center of the spiral (excluding the final element of the spiral)
        #and excluding the columns on the sides of the spiral
        for i in range (p,n-p):  #p1=0 k1=n  p in range
            b[p][i]=z
            z+=1
        c+=1
        
        #filling the columns on the right side exluding the  elements already filled in the previous loop
        for j in range (p+1,n-p): #p2=1 k2=n
            b[j][n-1-p]=z
            z+=1
        c+=1
        
        # #filling the rows from the bottom to the center of the spiral (excluding the last 4 values of the spiral)
        #and excluding the columns on the sides of the spiral
        for i in range (n-2-p,p-1,-1): #p3=n-2  k3=-1
            b[n-1-p][i]=z
            z+=1
        c+=1
        
        ##filling the columns on the left side exluding the  elements already filled in the previous loops
        for j in range (n-2-p,p,-1):  #p4=n-2   k4=0
            b[j][p]=z
            z+=1
        c+=1
    #the final elemnt of the spiral
    b[n//2][n//2]=z
    z+=1
    c+=1

#Everything made above adapted for the case when n is even. In this case, we fill the spiral in the same way as for an odd n,
#but inthe last step of the algorithm the final 4 values of the spiral are filled instead of 1 value for an odd n   
else:
   for p in range (0,a//4):
        for i in range (p,n-p):
            #p1=0 k1=n  p in range
            b[p][i]=z
            z+=1

        c+=1
        for j in range (p+1,n-p): #p2=1 k2=n
            b[j][n-1-p]=z
            z+=1
        c+=1
        for i in range (n-2-p,p-1,-1): #p3=n-2  k3=-1
            b[n-1-p][i]=z
            z+=1
        c+=1
        for j in range (n-2-p,p,-1):  #p4=n-2   k4=0
            b[j][p]=z
            z+=1
        c+=1
#last four values of the spiral

   b[n//2-1][n//2-1]=z
   z+=1
   c+=1

   b[n//2-1][n//2]=z
   z+=1
   c+=1

   b[n//2][n//2]=z
   z+=1
   c+=1

   b[n//2][n//2-1]=z
   z+=1
   c+=1

#Printing the list in a form of a spiral matrix  
for k in range (n):
    for m in range (n):
        if b[k][m]<10:
            print (b[k][m],end='   ')
        elif b[k][m]<100:
            print (b[k][m],end='  ') 
        else:
            print (b[k][m],end=' ')
    print()
```
